```python
import numpy as np
import pandas as pd
from pandas import read_csv as rc
```

```python
import matplotlib.pyplot as plt
```

```python
housing_data=rc(r"D:\yohesha\Code Casa Internship\Housing.csv")
```

```python
housing_data
```

```python
housing_data=pd.get_dummies(housing_data,columns=['mainroad','guestroom','basement','airconditioning','hotwaterheating','prefarea','furnishingstatus'],drop_first=True)
```

```python
housing_data.describe()
```

```python
housing_data
```

```python
from sklearn.preprocessing import MinMaxScaler
columns=['mainroad_yes','guestroom_yes','basement_yes','airconditioning_yes','hotwaterheating_yes','prefarea_yes','furnishingstatus_semi-furnished','furnishingstatus_unfurnished']
scaler = MinMaxScaler()
data=pd.DataFrame(scaler.fit_transform(housing_data),columns=housing_data.columns)
```

```python
data
```

```python
data_y=data[['price']]
```

```python
data_x=data.drop('price',axis=1)
```

```python
data_x
```

```python
data_y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(data_x,data_y,test_size=0.4,random_state=40)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
import math
math.sqrt(545)
#this is to find no of neighbors to be assigned
#This is calculated by taking the square root of the number of rows present in the entire dataset
```

```python
# Neighbor value = 23
n_neighbors=23
```

```python
knn_regressor=KNeighborsRegressor(n_neighbors)
```

```python
knn_regressor.fit(X_train,Y_train)
```

```python
knn_predictions=knn_regressor.predict(X_test)
```

```python
knn_predictions
```

```python
from sklearn.metrics import mean_squared_error
```

```python
MSE=mean_squared_error(Y_test,knn_predictions)
```

```python
MSE
```

```python
rmse=np.sqrt(MSE)
rmse
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test,knn_predictions)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(Y_test,knn_predictions)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR_housing.fit(X_train,Y_train)
```

```python
LR_predictions=LR_housing.predict(X_test)
```
